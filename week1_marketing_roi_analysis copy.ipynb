```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv("train_dataset.csv")

# List of amenities to analyze
amenities = ['coffee_bar', 'video_store', 'salad_bar', 'prepared_food', 'florist']

# Calculate average sales for each amenity (1 = has it, 0 = doesn't)
for amenity in amenities:
    avg_sales = df.groupby(amenity)['store_sales(in millions)'].mean().round(2)
    print(f"\nAverage sales by {amenity}:")
    print(avg_sales)


    # Plot
    avg_sales.plot(kind='bar' , figsize=(8,5), color=['lightcoral', 'teal'])
    plt.title(f'Average Store Sales by {amenity.replace("_", " ").title()} Presnce', fontsize=14)
    plt.ylabel("Average Sales (Millions $)")
    plt.xlabel(f'{amenity.replace("_", " ").title()} (0=No, 1=Yes)')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.savefig(f"sales_by_{amenity}.png" , dpi=300, bbox_inches='tight')
    plt.show()
print("\nAll plots saved!")

# Overall I think this is a weak analysis. You can't tell that much of a difference on these stores having an impact to sales based on these bar charts
# I will practice and come back to this to try and find better charts to compare
```
