```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('advertising.csv')

print('First 5 rows:')
print(df.head())

# Calculate ROI % for each channel
# ROI = (sales - spend) / spend * 100
df['TV_ROI_%'] = (df['sales'] - df['TV']) / df['TV'] * 100
df['Radio_ROI_%'] = (df['sales'] - df['radio']) / df['TV'] * 100
df['Newspaper_ROI_%'] = (df['sales'] - df['newspaper']) / df['TV'] * 100

# Average ROI by Channel
roi_summary = pd.DataFrame({'Channel': ['TV', 'Radio', 'Newspaper'], 'Average_ROI_%': [df['TV_ROI_%'].mean(), df['Radio_ROI_%'].mean(), df['Newspaper_ROI_%'].mean()]}).round(2)
print('\nAverage ROI % by Channel:')
print(roi_summary)

roi_summary.plot(x='Channel', y='Average_ROI_%', kind='bar', figsize=(10, 6), color=['teal', 'skyblue', 'salmon'])
plt.title('Average ROI % by Advertising Channel', fontsize=16)
plt.ylabel('ROI %')
plt.xlabel('Channel')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("roi_by_channel.png", dpi=300, bbox_inches='tight')
plt.show()

# Save summary
roi_summary.to_csv("roi_summary.csv", index=False)
print("\nPlot saved as roi_by_channel.png")
print("Summary saved as roi_summary.csv")
```
